```python
def L(u, v):
    return 3 * u**2 + 3 * v**2 - u*v + 7*u - 7*v + 10
def Lu(u, v):
    return 6* u - v + 7
def Lv(u, v):
    return 6* v - u - 7

u = np.linspace(-5, 5, 501)
v = np.linspace(-5, 5, 501)
U, V = np.meshgrid(u, v)
Z = L(U, V)
```

```python
# 勾配降下法のシミュレーション
W = np.array([4.0, 4.0])
W1 = [W[0]]
W2 = [W[1]]
N = 21
alpha = 0.05
for i in range(N):
    W = W - alpha *np.array([Lu(W[0], W[1]), Lv(W[0], W[1])])
    W1.append(W[0])
    W2.append(W[1])
```

```python
n_loop=11

WW1 = np.array(W1[:n_loop])
WW2 = np.array(W2[:n_loop])
ZZ = L(WW1, WW2)
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.set_zlim(0,250)
ax.set_xlabel('W')
ax.set_ylabel('B')
ax.set_zlabel('loss')
ax.view_init(50, 240)
ax.xaxis._axinfo["grid"]['linewidth'] = 2.
ax.yaxis._axinfo["grid"]['linewidth'] = 2.
ax.zaxis._axinfo["grid"]['linewidth'] = 2.
ax.contour3D(U, V, Z, 100, cmap='Blues', alpha=0.7)
ax.plot3D(WW1, WW2, ZZ, 'o-', c='k', alpha=1, markersize=7)
plt.show()
```

```python
sampleData1 = np.array([
    [166, 58.7],
    [176.0, 75.7],
    [171.0, 62.1],
    [173.0, 70.4],
    [169.0, 60.1]
])
print(sampleData1)
```

```python
# 身長だけ抜き出したものと体重だけ抜き出したものを用意する
x = sampleData1[:,0] #全ての行の1列目、身長
y = sampleData1[:,1] #全ての行の2列目、体重
```

```python
plt.scatter(x, y, c='k', s=50)
plt.xlabel('$x$:身長(cm)')
plt.ylabel('$y$:体重(kg)')
plt.title('身長と体重の関係')
plt.show()
```

```python
X = x - x.mean()
Y = y - y.mean()
```

```python
plt.scatter(X, Y, c='k', s=50)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('加工後の身長と体重の関係')
plt.show()
```

```python
# XとY（NumPy）をテンソル変数化する
X = torch.tensor(X).float()
Y = torch.tensor(Y).float()
```

```python
print(X)
print(Y)
```

```python
# パラメータの定義
# WとBは勾配計算をするので、自動微分機能をTrueにする
W = torch.tensor(1.0, requires_grad=True).float()
B = torch.tensor(1.0, requires_grad=True).float()
```

```python
# 予測関数は1次関数
def pred(x):
    return W * X + B
```

```python
Yp = pred(x)
print(Yp)
```

```python
# 予測値の計算グラフ表示
params = {'W':W,'B':B}
g = make_dot(Yp, params=params)
display(g)
```

```python
# 損失計算
def mse(Yp, Y):
    loss = ((Yp - Y) ** 2).mean()
    return loss
```

```python
loss = mse(Yp, Y)
print(loss)
```

```python
params = {'W':W, 'B':B}
g = make_dot(loss, params=params)
display(g)
```

```python
loss.backward()
```

```python
print(W.grad)
print(B.grad)
```

```python
# 学習率の定義
lr = 0.001
```

```python
# 勾配を元にパラメータ修正
# with torch.no_grad() を付ける必要がある

with torch.no_grad():
    W -= lr * W.grad
    B -= lr * B.grad
    
    # 計算済みの勾配値をリセットする
    W.grad.zero_()
    B.grad.zero_()
```

```python
# パラメータと勾配値の確認
print(W)
print(B)
print(W.grad)
print(B.grad)
```

```python
# 初期化
W = torch.tensor(1.0, requires_grad=True).float()
B = torch.tensor(1.0, requires_grad=True).float()
# 繰り返しの回数
num_epochs = 500
# 学習率
lr = 0.001
# 記録用配列初期化
history = np.zeros((0, 2))
```

```python
# ループ処理
for epoch in range(num_epochs):
    # 予測計算
    Yp = pred(X)
    # 損失計算
    loss = mse(Yp, Y)
    # 勾配計算
    loss.backward()
    
    with torch.no_grad():
        W -= lr * W.grad
        B -= lr * B.grad
        # 勾配の初期化
        W.grad.zero_()
        B.grad.zero_()

    # 10回に1回記録する
    # 損失の記録
    if (epoch %10 == 0):
        item = np.array([epoch, loss.item()])
        history = np.vstack((history, item))
        print(f'epoch = {epoch}  loss = {loss:.4f}')
```

```python
# パラメータの最終値
print('W = ', W.data.numpy())
print('B = ', B.data.numpy())

#損失の確認
print(f'初期状態: 損失:{history[0,1]:.4f}') 
print(f'最終状態: 損失:{history[-1,1]:.4f}') 
```

```python
plt.plot(history[:,0], history[:,1],'b')
plt.xlabel('繰り返し回数')
plt.ylabel('損失')
plt.title('学習曲線(損失)')
plt.show()
```

```python
# ループ処理

for epoch in range(num_epochs):

    # 予測計算
    Yp = pred(X)
    
    # 損失計算
    loss = mse(Yp, Y)

    # 勾配計算
    loss.backward()

    # パラメータ修正
    optimizer.step()
    
    #勾配値初期化
    optimizer.zero_grad()
        
    # 損失値の記録
    if (epoch %10 == 0):
        item = np.array([epoch, loss.item()])
        history = np.vstack((history, item))
        print(f'epoch = {epoch}  loss = {loss:.4f}')
```

```python
# パラメータの最終値
print('W = ', W.data.numpy())
print('B = ', B.data.numpy())

#損失の確認
print(f'初期状態: 損失:{history[0,1]:.4f}') 
print(f'最終状態: 損失:{history[-1,1]:.4f}') 
```

```python
# 学習曲線の表示 (損失)

plt.plot(history[:,0], history[:,1], 'b')
plt.xlabel('繰り返し回数')
plt.ylabel('損失')
plt.title('学習曲線(損失)')
plt.show()
```

```python
# 初期化

# WとBを変数として扱う
W = torch.tensor(1.0, requires_grad=True).float()
B = torch.tensor(1.0, requires_grad=True).float()

# 繰り返し回数
num_epochs = 500

# 学習率
lr = 0.001

# optimizerとしてSGD(確率的勾配降下法)を指定する
import torch.optim as optim
optimizer = optim.SGD([W, B], lr=lr, momentum=0.9)

# 記録用配列初期化
history2 = np.zeros((0, 2))
```

```python
# ループ処理

for epoch in range(num_epochs):

    # 予測計算
    Yp = pred(X)
    
    # 損失計算
    loss = mse(Yp, Y)

    # 勾配計算
    loss.backward()

    # パラメータ修正
    optimizer.step()
    
    #勾配値初期化
    optimizer.zero_grad()
        
    # 損失値の記録
    if (epoch %10 == 0):
        item = np.array([epoch, loss.item()])
        history2 = np.vstack((history2, item))
        print(f'epoch = {epoch}  loss = {loss:.4f}')
```

```python
# 学習曲線の表示 (損失)

plt.plot(history[:,0], history[:,1], 'b', label='デフォルト設定')
plt.plot(history2[:,0], history2[:,1], 'k', label='momentum=0.9')
plt.xlabel('繰り返し回数')
plt.ylabel('損失')
plt.legend()
plt.title('学習曲線(損失)')
plt.show()
```
